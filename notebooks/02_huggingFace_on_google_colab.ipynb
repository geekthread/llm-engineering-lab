## Stages & Concepts

### Stage 1 — Cloud GPU Playground
Verify the GPU environment, then explore two non-text modalities to get comfortable
with the HuggingFace ecosystem before touching language models.

| Sub-section | Model Used | What You Learn |
|-------------|-----------|----------------|
| 1.1 Text-to-Image | `stabilityai/sdxl-turbo` | How diffusion models denoise random noise guided by a text prompt; 4-step fast inference |
| 1.2 Text-to-Speech | `microsoft/speecht5_tts` | How speaker embeddings steer voice identity; mel-spectrogram → waveform pipeline |

### Stage 2 — HuggingFace Pipelines: AI in One Line
The high-level inference API that handles model loading, tokenisation, post-processing,
and device placement automatically. Six NLP tasks, each in a handful of lines.

| Sub-section | Task | Model Used | What You Learn |
|-------------|------|-----------|----------------|
| 2.1 | Sentiment Analysis | `distilbert-base-uncased-finetuned-sst-2-english` + `nlptown/bert-base-multilingual-uncased-sentiment` | Swapping models with zero API change; multilingual support |
| 2.2 | Named Entity Recognition | Default NER model | Entity extraction and aggregation strategy |

### Stage 3 — Tokenizers: The Rosetta Stone of LLMs
The crucial insight most tutorials skip: LLMs read integers, not words.

| Sub-section | What You Learn |
|-------------|----------------|
| 3.0 Vocabulary & Encoding | How `tokenizer.encode()` converts strings to token ID sequences; character vs word vs token count |
| 3.0 BPE | Byte-Pair Encoding — how vocabularies are built by merging frequent byte pairs |
| 3.0 Special tokens | `get_added_vocab()` — the structural glue of the chat format |
| 3.1 Chat Templates | `apply_chat_template()` — how messages=[...] becomes a token sequence; why Llama and Phi-4 templates differ completely |
| 3.2 Code Tokenisation | Why QwenCoder uses fewer tokens for code than a general tokenizer; vocabulary extension for programming constructs |

**Models compared:** `meta-llama/Meta-Llama-3.1-8B-Instruct`, `microsoft/Phi-4-mini-instruct`, `Qwen/Qwen2.5-Coder-7B-Instruct`

---

### Stage 4 — Direct Model Control: Under the Hood

| Sub-section | What You Learn |

| 4.3 Multi-model comparison | Same prompt across five models: Llama 3.2-1B, Phi-4 mini, Gemma 270M, Qwen3-4B, DeepSeek-R1-Distill |

## Key Concepts at a Glance

| Concept | One-liner |
|---------|-----------|
| **Tokenization** | Models see integers, not words |
| **BPE** | Vocabulary built by merging frequent byte pairs |
| **Chat template** | Each model family uses unique special tokens |
| **4-bit Quantization (NF4)** | 4× less GPU memory, ~5% accuracy cost |
| **Double quantization** | Further compresses quantization scale factors |
| **Streaming** | `TextStreamer` prints tokens as they are generated |
| **Pipeline API** | High-level API; handles all tokenisation and device plumbing |
| **Whisper ASR** | Encoder-decoder transformer for speech-to-text |
| **Prompt engineering** | System prompt structure directly controls output format |
| **Diffusion** | Image generation by iterative noise removal |
| **Speaker embedding** | Vector encoding of voice identity for TTS |
